#Arbol de decision

```python
# Importamos libreria
import pandas as pd

# Importamos DataFrame

pacientes = pd.read_csv('/content/Pacientes2.csv', index_col=0)
```

```python
pacientes.head()
```

#analisis de datos

```python
pacientes.info()
```

##separamos las columnas de variables predictoras, de la columna que contienen la variable a predecir

```python
#Variables predictoras
x = pacientes.iloc[:,1:11]

#Variables a predecir
y = pacientes.iloc[:,0]

#Mostramos los datos
x.head()
```

##Dividimos datos para entrenamiento y prueba
generalmente se ocupan un 75% - 80% datos a entrenamiento

```python
from sklearn.model_selection import train_test_split

#x_train , y_train = datos para entrenamiento
#x_test , y_test = datos para la prueba del modelo
x_train, x_test, y_train,y_test = train_test_split(x,y, train_size=0.75, random_state=0) #random_state indica que siempre ocuparemos las mismas filas de entrenamiento
```

```python
x_train.info()
```

# creacion de modelo

```python
from sklearn.tree import DecisionTreeClassifier

#Construccion de arbol de decision

arbol = DecisionTreeClassifier(max_depth=4)

#Entrenamiento del arbol de decision
arbol_enfermedad = arbol.fit(x_train,y_train)
```

#Grafico del Modelo Arbol
analisis:
por el lado izq es correcto o true, y por el lado derecho es False

```python
from matplotlib import pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(25,25)) # parametros del grafico

tree.plot_tree (arbol_enfermedad, feature_names = list (x.columns.values), 
                class_names = list (y.values), filled=True)

plt.show()
```

#prediccion del modelo
utilizar los datos de la tabla testing

```python
y_pred = arbol_enfermedad.predict(x_test)
y_pred
```

#calculo de prediccion

```python
from sklearn.metrics import confusion_matrix
import numpy as np


# diagonal principal datos correctos
# lo demas datos incorrectos
matriz = confusion_matrix(y_test, y_pred)
matriz
```

calculo de precision global

```python
precision = np.sum(matriz.diagonal())/np.sum(matriz)
precision
```

calculo de precision por cada clase

```python
precisionclase_no = ((matriz[0,0]))/sum(matriz[0,])
precisionclase_no                                        
```

```python
precisionclase_si = ((matriz[1,1]))/sum(matriz[1,])
precisionclase_si                                        
```
